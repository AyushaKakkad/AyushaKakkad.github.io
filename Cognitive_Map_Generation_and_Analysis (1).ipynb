####Function to translate Arabic Interview Data to English

```python
translate = Translator()

def translate_to_english(text):
    translation = translate.translate(text, src='ar', dest='en')
    return translation.text
```

####Translate the Arabic Interview data to English

```python
# Translate the Arabic interview data to English
translated_interviews = [translate_to_english(interview) for interview in ara_interview]

# Print the translated Arabic interview text
for i, translated_text in enumerate(translated_interviews, start=1):
    print(f"Translated Arabic Interview {i}: {translated_text}")
```

####Combine Interview Data

```python
combined_interviews = translated_interviews + eng_interview
```

####Preprocess Data

```python
nlp = spacy.load('en_core_web_sm')

def preprocess_text(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
    return ' '.join(tokens)

processed_interviews = [preprocess_text(interview) for interview in combined_interviews]
```

####Tokenize Data

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(processed_interviews)
```

####Topic Modelling to identify Key Entities

```python
lda = LatentDirichletAllocation(n_components=2, random_state=123)
lda.fit(X)

# Get the topics and their corresponding words
topics = lda.components_
feature_names = vectorizer.get_feature_names_out()

topic_assignments = lda.transform(X)
labels = topic_assignments.argmax(axis=1)

# Extract the key entities and their relationships
key_entities = []
for topic_idx, topic in enumerate(topics):
    key_entities.append([feature_names[i] for i in topic.argsort()[:-6:-1]])

print("Key Entities for each topic:")
for idx, entities in enumerate(key_entities):
    print(f"Topic {idx+1}: {entities}")
```

####Generate Cognitive Map using NetworkX

```python
cog_map = nx.DiGraph()

    # Add nodes and edges based on key entities
for idx, entity in enumerate(key_entities):
      for i in range(len(entity) - 1):
            cog_map.add_edge(entity[i], entity[i+1], weight=idx+1)
```

####Plot Cognitive Map using Matplotlib

```python
plt.figure(figsize=(12, 10))

#Position nodes of graph using the spring layout
pos = nx.spring_layout(cog_map, k=0.5, iterations=100)

#Draw nodes
nx.draw_networkx_nodes(cog_map, pos, node_size=3500, node_color='lightblue')

#Draw edges with weights showing strength of relationships between 2 nodes
edges = nx.draw_networkx_edges(cog_map, pos, arrowstyle='-|>', arrowsize=20, edge_color='gray', width = 2)
edge_labels = nx.get_edge_attributes(cog_map, 'weight')
nx.draw_networkx_edge_labels(cog_map, pos, edge_labels=edge_labels, font_color='red', font_size=10)

#Draw node labels
nx.draw_networkx_labels(cog_map, pos, font_size=12, font_weight='bold', font_color='black')

# Add title
plt.title('Cognitive Map of Interview Data', fontsize=16, fontweight='bold')

# Remove axis
plt.axis('off')

# Display the plot
plt.show()
```

####Cognitive Map Analysis

```python
# Function to identify central concepts based on node degrees
def identify_central_concepts(cog_map):
    degrees = dict(cog_map.degree())
    sorted_degrees = sorted(degrees.items(), key=lambda item: item[1], reverse=True)
    return sorted_degrees

# Function to examine relationships by inspecting edge weights
def examine_relationships(cog_map):
    edges = cog_map.edges(data=True)
    sorted_edges = sorted(edges, key=lambda x: x[2]['weight'], reverse=True)
    return sorted_edges

# Function to analyze clusters using community detection
def analyze_clusters(cog_map):
    from networkx.algorithms.community import greedy_modularity_communities
    communities = list(greedy_modularity_communities(cog_map))
    return communities

# Perform the analysis
central_concepts = identify_central_concepts(cog_map)
relationships = examine_relationships(cog_map)
clusters = analyze_clusters(cog_map)

# Convert central entity to DataFrame for better visualization
central_concepts_df = pd.DataFrame(central_concepts, columns=['Entity', 'Degree'])

# Print central concepts
print("Central Concepts:")
print(central_concepts_df)

# Convert relationships to DataFrame for better visualization
relationships_df = pd.DataFrame(relationships, columns=['Belief Classes', 'Decision Classes', 'Connections'])

# Print relationships
print("\nRelationships:")
print(relationships_df)

# Print clusters
print("\nClusters:")
for idx, cluster in enumerate(clusters, start=1):
    print(f"Cluster {idx}: {list(cluster)}")

# Visualize the cognitive map with community clusters
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(cog_map, k=0.5, iterations=100)
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink']

# Draw nodes with colors based on clusters
for i, cluster in enumerate(clusters):
    nx.draw_networkx_nodes(cog_map, pos, nodelist=list(cluster), node_size=3500, node_color=colors[i % len(colors)])

# Draw edges with customized settings and labels
nx.draw_networkx_edges(cog_map, pos, arrowstyle='-|>', arrowsize=20, edge_color='gray')
edge_labels = nx.get_edge_attributes(cog_map, 'weight')
nx.draw_networkx_edge_labels(cog_map, pos, edge_labels=edge_labels, font_color='red', font_size=10)

# Draw node labels with customized settings
nx.draw_networkx_labels(cog_map, pos, font_size=12, font_weight='bold', font_color='black')

# Add a title with customized settings
plt.title('Cognitive Map of Interview Data with Clusters', fontsize=16, fontweight='bold')

# Remove axis
plt.axis('off')

# Display the plot
plt.show()
```
